# Selecting our datasets

```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
jogging = pd.read_csv("./A3_dataset/Jogging.csv")
cycling = pd.read_csv("./A3_dataset/Cycling.csv")
football = pd.read_csv("./A3_dataset/Football.csv")
jumprope = pd.read_csv("./A3_dataset/JumpRope.csv")
pushups= pd.read_csv("./A3_dataset/Pushups.csv")
sitting = pd.read_csv("./A3_dataset/Sitting.csv")
swimming = pd.read_csv("./A3_dataset/Swimming.csv")
tennis = pd.read_csv("./A3_dataset/Tennis.csv")
walking = pd.read_csv("./A3_dataset/Walking.csv")
```

```python
activities = [jogging, cycling, football, jumprope, pushups, sitting, swimming, tennis, walking]
strings = ["jogging", "cycling", "football", "jumprope", "pushups", "sitting", "swimming", "tennis", "walking"]


count = 0
for activity in activities:
    # date = activity["YYYY-MO-DD HH-MI-SS_SSS"][0]
    dt_obj = datetime.strptime("14.5.2021 13:47:51,11",'%d.%m.%Y %H:%M:%S,%f')
    millisec = int(dt_obj.timestamp() * 1000)
    timestamps_ms = []
    for i in range(len(activity)):
        if "Time since start in ms " in activity.columns:
            timestamps_ms.append(activity["Time since start in ms "][i] + millisec)
        elif "Time since start in ms" in activity.columns:
            timestamps_ms.append(activity["Time since start in ms"][i] + millisec)
        
    activity["Timestamps (ms)"] = timestamps_ms
    # activity["activity"] = strings[activities.index(activity)]
    activity["activity"] = strings[count]

    count += 1
```

```python
# The tennis dataset had different column names
tennis.columns = jogging.columns
```

```python
combined_dataset = pd.concat(activities, axis = 0)
```

```python
combined_dataset.to_csv("./A3_dataset/COMBINED.csv", index=False)
```
